```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

import datetime

num_rows = 5
num_cols = 4

current_datetime = datetime.datetime.now()
timestamp_str = current_datetime.strftime('%Y-%m-%d')
owner ='chuangm'

# plt.title('mIOU curve')
fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 20))
index = [1,3,5,12,18]

for i in range(num_rows):
    img_0 = Image.open("/content/gdrive/MyDrive/patterm/hw3/hw3_semantic_segmentation/results/"+str(index[i])+"_image.png")
    axs[i, 0].imshow(img_0)
    axs[i, 0].axis('off')

    img_1 = Image.open("/content/gdrive/MyDrive/patterm/hw3/hw3_semantic_segmentation/results/"+str(index[i])+"_target.png")
    axs[i, 1].imshow(img_1)
    axs[i, 1].axis('off')

    img_1 = Image.open("/content/gdrive/MyDrive/patterm/hw3/hw3_semantic_segmentation/results/"+str(index[i])+"_pred.png")
    axs[i, 2].imshow(img_1)
    axs[i, 2].axis('off')

    img_1 = Image.open("/content/gdrive/MyDrive/patterm/hw3/hw3_semantic_segmentation/results/"+str(index[i])+"_pred_.png")
    axs[i, 3].imshow(img_1)
    axs[i, 3].axis('off')

plt.suptitle(f'Result comparing\n {owner}: {timestamp_str}', fontsize=16)
axs[0, 0].set_title("RGB Image")
axs[0, 1].set_title("Ground Truth annotation")
axs[0, 2].set_title("Model 1 prediction")
axs[0, 3].set_title("Model 2 prediction")
plt.tight_layout()

plt.show()
```
